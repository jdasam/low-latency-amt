```python
path = 'data/MAESTRO/'
metadata = json.load(open(os.path.join(path, 'maestro-v3.0.0.json')))
metadata_df = pd.DataFrame(metadata)
files = sorted([(os.path.join(path, metadata_df.iloc[idx]['audio_filename'].replace('.wav', '.flac')),
                  os.path.join(path, metadata_df.iloc[idx]['midi_filename'])) for idx in range(len(metadata_df)) if metadata_df.iloc[idx]['split'] == 'train'])

files = [(audio if os.path.exists(audio) else audio.replace('.flac', '.wav'), midi) for audio, midi in files]
```
